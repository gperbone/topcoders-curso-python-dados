### 3. Junte as colunas no seu dataframe de items necessarias para poder ver quais são as 10 categorias que mais trouxeram receita pra a empresa

```python
df_categorias = products_df[['product_id','product_category_name']]
df_precos = items_df[['product_id','price']]

df_final = pd.merge(df_categorias,df_precos, how = 'left', on = 'product_id')

df_final.groupby(by = 'product_category_name')['price'].sum().nlargest(10)
```

```python
#número de itens no maior pedido
display(items_df['order_item_id'].max())

#order_id do maior pedido
codigo_mais_itens = items_df[items_df.order_item_id == items_df.order_item_id.max()]['order_id'].iloc[0]
codigo_mais_itens
```

### 5. No DF de orders crie uma nova coluna com o time delta entre a data estimada de entrega e a data real 

##### link strptime()
[link strptime()](https://www.programiz.com/python-programming/datetime/strptime)

```python
from datetime import datetime, timedelta

orders_df['order_delivered_customer_date'] = orders_df['order_delivered_customer_date'].apply(lambda x : datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S') if type(x) in [str] else x)
orders_df['order_estimated_delivery_date'] = orders_df['order_estimated_delivery_date'].apply(lambda x : datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S') if type(x) in [str] else x)
orders_df.info()

#outro modo
#orders_df['order_estimated_delivery_date'] = pd.to_datetime(orders_df['order_estimated_delivery_date'], format = '%Y-%m-%d %H:%M:%S')
```

```python
orders_df["tempo_de_entrega"] = orders_df['order_delivered_customer_date'] - orders_df['order_estimated_delivery_date']
orders_df.head()
```

### 6. Analise a coluna e traga conclusões
    - temos valores outliers?
    - quanto eles representam?
    - qual é a nossa media?
    - quanto representa de nosso dataframe pedidos que demoraram mais de 5 dias para entregar apos a data prometida

```python
#calculando o intervalo de outliers
iqr = orders_df.describe()['tempo_de_entrega'].iloc[6] - orders_df.describe()['tempo_de_entrega'].iloc[4] 
print(iqr)
limite_sup = iqr*1.5 + orders_df.describe()['tempo_de_entrega'].iloc[6]
print(limite_sup)
limite_inf = -iqr*1.5 + orders_df.describe()['tempo_de_entrega'].iloc[4]
print(limite_inf)

filtro_outlier_sup = orders_df['tempo_de_entrega'] > limite_sup
filtro_outlier_inf = orders_df['tempo_de_entrega'] < limite_inf

# porcentagem de outliers
print('A porcentagem é de:', orders_df[filtro_outlier_sup | filtro_outlier_inf].shape[0]/orders_df.shape[0]*100)

#média de tempo de entrega
print('A média é de tempo para entrega:', orders_df.describe()['tempo_de_entrega'].iloc[1])

# pedidos com atraso de 5 ou mais dias
x = timedelta(days=5)
filter_outlier_5_days = orders_df['tempo_de_entrega'] > x

print('Numero de pedidos atrasados 5 dias ou mais:', orders_df[filter_outlier_5_days].shape[0])
```

### 7. Quantos estados temos no nosso dataset? Tem todos os estados do Brasil? Quais são os 3 estados com mais pedidos?

```python
estados_no_df = customers_df['customer_state'].unique()
display(estados_no_df)
 
display(customers_df['customer_state'].nunique())

orders_state = orders_df.merge(customers_df[:][['customer_id','customer_state']], how = 'left')

orders_state.groupby(by = 'customer_state')['order_id'].count().nlargest(3)
```

8. Bonus(Faça um gráfico com essa informação)

```python
import matplotlib.pyplot as plt

orders_per_state = orders_state.groupby(by = 'customer_state')['order_id'].count().sort_values(ascending=False).reset_index()
display(orders_per_state)

plt.figure(figsize=(20,10))

plt.bar(orders_per_state['customer_state'], orders_per_state['order_id'])
```
