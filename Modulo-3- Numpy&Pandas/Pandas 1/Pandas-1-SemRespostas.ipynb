#### Series 

```python
#Pandas Series
#Um grupo de series forma um dataframe
ds1 = pd.Series(np.random.randint(10, 25, size=5))
ds2 = pd.Series(np.random.randint(10, 25, size=5))
print(type(ds1))
print(ds1.shape)
print(ds1)
print('---------')
print(type(ds2))
print(ds2.shape)
print(ds2)
```

```python
#Matematica
ds = ds1 + ds2
print('Soma:')
print(ds)
print('\n')
print('Subtração:')
ds = ds1 - ds2
print(ds)
print('\n')
print('Multiplicação:')
ds = ds1 * ds2
print(ds)
print('\n')
print("Divisão:")
ds = ds1 / ds2
print(ds)
print('\n')
print('Media:')
print(ds1.mean())
print('\n')
print('Mediana:')
print(ds1.median())
print('\n')
print('Std Dev:')
print(ds1.std())
print('\n')
print('\n')
```

```python
#Comparando series
print("Igual:")
print(ds1 == ds2)
print("Maior que:")
print(ds1 > ds2)
print("Menor que:")
print(ds1 < ds2)
display(ds1[ds1 < ds2],ds2[ds1<ds2])
```

##### DYI Series

- 1 Crie duas series aleatorias com 30 numeros, no intervalo 20 a 50 cada  ds1 ds2
- 2 Crie uma nova serie com o nome g_than com a comparação entre a s1 e a s2 
- 3 Crie uma nova serie com o nome ds_sum com a soma das duas series 
- 4 Apresente o desvio padrão da serie ds_sum
- 5 Crie uma nova serie com 30 strings de 5 caracteres aleatorios (Dica. random.choice + string.ascii_lowercase)
- 6 Crie uma lista com todas as series criadas 
- 7 Crie um dataframe com as suas series (Dica. pd.concat())
- 8 Coloque o nome de cada serie no novo dataframe (Dica .rename)

```python
import random 
import string

#1 
ds1 = pd.Series(np.random.randint(20, 50, size=30))
ds2 = pd.Series(np.random.randint(20, 50, size=30))
```

```python
#2 gthan
g_than = ds1 > ds2
# print(g_than)

#3
ds_sum = ds1 + ds2
# print(ds_sum)

#4
ds_desv = ds_sum.std()
# print(ds_desv)

#5
ds_string = pd.Series([''.join(random.choices(string.ascii_lowercase, k=5)) for x in range(30)])
# print(ds_string)

#6
list_series = [ds1, ds2, g_than, ds_sum, ds_string]
# print(my_series)
# print(type(list_series))

#7
datasrs = pd.concat(list_series, axis = 1)
# datasrs

#8
datasrs.rename(columns={0:'ds1', 1: 'ds2', 2:'g_than', 3:'ds_sum', 4:'ds_string'}, inplace=True)
datasrs
```

```python
#Pandas as Type
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.astype.html
df_csv['Valor de Venda'] = df_csv['Valor de Venda'].apply(lambda x:x.replace(',','.'))

df_csv['Valor de Venda'] = df_csv['Valor de Venda'].astype('float')
```

>Uma função lambda do Python se comporta como uma função normal em relação aos argumentos. Portanto, um parâmetro lambda pode ser inicializado com um valor padrão: o parâmetro **n** toma o **n** externo como valor padrão.

```python
df_csv['Numero Rua'] = df_csv['Numero Rua'].apply(lambda x:0 if x == 'S/N' or 'SN' else x)
#As Type mais de uma coluna
df_csv = df_csv.astype({"Valor de Venda":float,"Numero Rua":int})
```
